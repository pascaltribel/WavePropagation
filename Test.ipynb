```python
from WaveDataset import *
import torch
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import sklearn.decomposition
```

```python
dataset = WaveDataset(num_samples=10, nx=32, speed=0.005, dt=0.01, t=5, interval=32**2, sampling_rate=1, verbose=True)
```

```python
dataset.flattened = True
```

```python
evol = dataset[0][2]
```

```python
pca = sklearn.decomposition.PCA(n_components=8)
a = pca.fit(evol)
```

```python
b = pca.inverse_transform(pca.transform(dataset[0][2]))
```

```python
b = b.reshape((32**2, 32, 32))
```

```python
for i in b:
    plt.imshow(i, vmin=np.min(b), vmax=np.max(b))
    plt.show()
```
